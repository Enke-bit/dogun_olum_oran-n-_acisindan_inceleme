### Model kurma

##### 1. Model kurma

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
```

```python
# Türkiye verisini seçme
turkey_data = df[df['Country Name'] == 'Turkiye']

# 'Years' sütununu zaman serisi indeksi olarak ayarlama
turkey_data.set_index('Years', inplace=True)

# Özellikler ve hedef değişkenleri ayırma
features = turkey_data[['Fertility Rate', 'GDP Growth', 'Gnipc', 'HDI', 'MYS']]
target = turkey_data['Death Rate']

# Veriyi normalleştirme
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
scaled_target = scaler.fit_transform(target.values.reshape(-1, 1))

# Veriyi train ve test setlerine ayırma
split_index = int(len(turkey_data) * 0.8)
X_train, X_test = scaled_features[:split_index], scaled_features[split_index:]
y_train, y_test = scaled_target[:split_index], scaled_target[split_index:]

# Veriyi LSTM için uygun formata getirme
def create_sequences(data, target, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(target[i+seq_length])
    return np.array(X), np.array(y)

SEQ_LENGTH = 5
X_train_seq, y_train_seq = create_sequences(X_train, y_train, SEQ_LENGTH)
X_test_seq, y_test_seq = create_sequences(X_test, y_test, SEQ_LENGTH)

# LSTM Modeli oluşturma
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(SEQ_LENGTH, X_train_seq.shape[2])))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Modeli eğitme
model.fit(X_train_seq, y_train_seq, epochs=50, batch_size=32, validation_split=0.2)

# Modeli değerlendirme
y_pred = model.predict(X_test_seq)
y_test_inv = scaler.inverse_transform(y_test_seq)
y_pred_inv = scaler.inverse_transform(y_pred)

# Performansı değerlendirme
mse = mean_squared_error(y_test_inv, y_pred_inv)
print(f"Mean Squared Error: {mse}")
```

##### 2. Modele tahmin yaptırma.

```python
from sklearn.preprocessing import MinMaxScaler

# Eğitim veri seti için özellikler ölçekleyici oluşturma ve eğitme
feature_scaler = MinMaxScaler()
feature_scaler.fit(features)  # Eğitim veri seti ile fit et

# Özellikler ve hedef değişkenler için ölçekleme
scaled_features = feature_scaler.transform(features)
target_scaler = MinMaxScaler()
scaled_target = target_scaler.fit_transform(target.values.reshape(-1, 1))
```

```python
# Türkiye verisini seçme
new_turkey_data = df[df['Country Name'] == 'Turkiye']

# 'Years' sütununu zaman serisi indeksi olarak ayarlama
new_turkey_data.set_index('Years', inplace=True)

# Özellikler ve hedef değişkenleri ayırma
new_features = new_turkey_data[['Fertility Rate', 'GDP Growth', 'Gnipc', 'HDI', 'MYS']]
new_target = new_turkey_data['Death Rate']

# Özellikler için eğitim veri seti ile fit edilmiş scaler'ı kullanarak normalleştirme
scaled_new_features = feature_scaler.transform(new_features)
scaled_new_target = target_scaler.transform(new_target.values.reshape(-1, 1))

# Veriyi LSTM için uygun formata getirme
X_new_seq, y_new_seq = create_sequences(scaled_new_features, scaled_new_target, SEQ_LENGTH)

# Tahmin yapma
y_new_pred = model.predict(X_new_seq)
y_new_target_inv = target_scaler.inverse_transform(y_new_seq)
y_new_pred_inv = target_scaler.inverse_transform(y_new_pred)

# Tahminler ile gerçek değerleri yan yana yazdırma
results = pd.DataFrame({
    'Real Death Rate': y_new_target_inv.flatten(),
    'Predicted Death Rate': y_new_pred_inv.flatten()
})

print(results)
```

##### 3. Tahmin sonuçlarını görselleştirme.

```python
# Yılları veya zaman dilimlerini almak
years = new_turkey_data.index[SEQ_LENGTH:]  # Seq_length kadar kaydırılmış index

plt.figure(figsize=(12, 6))

# Gerçek ve tahmin edilen değerleri çizme
plt.plot(years, y_new_target_inv.flatten(), label='Real Death Rate', color='blue')
plt.plot(years, y_new_pred_inv.flatten(), label='Predicted Death Rate', color='red', linestyle='--')

plt.xlabel('Years')
plt.ylabel('Death Rate')
plt.title('Real vs Predicted Death Rate')
plt.legend()
plt.xticks(rotation = 90)
plt.grid(True)
plt.show()
```
